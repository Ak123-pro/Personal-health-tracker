```python
import pandas as pd

food_df = pd.read_csv("food_nutrition.csv")

food_df.columns = food_df.columns.str.strip()

food_log = []

total_calories = 0
total_protein = 0
total_carbs = 0
total_fat = 0

print("\nAVAILABLE FOODS:")

print(food_df["food_name"].to_string(index=False))

while True:
    print("\n-----ADD FOOD INTAKE-----")
    food_name = input("Enter food name exactly as shown from the list")

    try:
        quantity = float(input("Enter quantity eaten in grams):"))
    except ValueError:
        print("please enter a valid number for quantity.")
        continue
    selected_food = food_df[food_df["food_name"]==food_name]

    if selected_food.empty:
        print("Food not found!.Please check spelling.:")
        continue
    
    calories = (selected_food["calories_per_100g"].values[0] / 100) * quantity
    protein = (selected_food["protein_g"].values[0] / 100) * quantity
    carbs = (selected_food["carbs_g"].values[0] / 100) * quantity
    fat = (selected_food["fat_g"].values[0] / 100) * quantity


    food_log.append({
        "Food": food_name,
        "Quantity (g)": quantity,
        "Calories (kcal)": round(calories, 2),
        "Protein (g)": round(protein, 2),
        "Carbs (g)": round(carbs, 2),
        "Fat (g)": round(fat, 2)
    })

    total_calories += calories
    total_protein += protein
    total_carbs += carbs
    total_fat += fat

    print("\n Food added successfully")
    print(f"Calories:{calories:.2f} kcal")

    more =input("Do you want to add more food? (yes/no):").strip().lower()
    if more!="yes":
        break

print("\n===== FOOD-WISE NUTRITION DETAILS =====")
for item in food_log:
    print(
        f"{item['Food']} ({item['Quantity (g)']} g) → "
        f"{item['Calories (kcal)']} kcal"
    )

print("\n=========DAILY NUTRITION SUMMARY=========")


print(f"Total Calories: {total_calories:.2f} kcal")
print(f"Total Protein: {total_protein:.2f} g")
print(f"Total Carbs: {total_carbs:.2f} g")
print(f"Total Fat: {total_fat:.2f} g")

#give intructions for better food habbit like i want to reduce carb in my diet what to do
# give info about the water intake
# add more food items also make convinent of entering food items names
# prepare a dataset of food with more info like medical 

```
